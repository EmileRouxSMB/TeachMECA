```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
%matplotlib nbagg
```

# Trajectoire d'un projectille 
## Données

```python
alpha = np.deg2rad(45)
g  = 10 #m/s**2
V0 = 5.
gamma = -g
```

## Trajectoire pour un alpha donnée

```python
V0x = V0*np.cos(alpha)
V0y = V0*np.sin(alpha)
```

```python
t=np.linspace(0,1,100)
Vx = V0x * np.ones_like(t)
Vy = V0y * np.ones_like(t)
```

```python
x  = V0x*t
y  = V0y * t + 0.5*gamma*t**2
```

```python
plt.figure()
plt.plot(x,y)
plt.grid('on')
plt.xlabel('x[m]')
plt.ylabel('y[m]')
```

## Généralisation en fonction de alpha

```python
def calc_traj(alpha):
    alpha = np.deg2rad(alpha)
    V0x = V0*np.cos(alpha)
    V0y = V0*np.sin(alpha)
    t=np.linspace(0,1,100)
    Vx = V0x * np.ones_like(t) + 0.
    Vy = V0y * np.ones_like(t) + gamma*t
    x  = V0x*t
    y  = V0y * np.ones_like(t)*t + 0.5*gamma*t**2
    return x,y
    
```

```python
plt.figure()
alphas = np.linspace(0,90,30)
for alpha in alphas:
    x,y = calc_traj(alpha)  
    plt.plot(x,y)
    
plt.grid('on')
plt.xlabel('x[m]')
plt.ylabel('y[m]')
```
